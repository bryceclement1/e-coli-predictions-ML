```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/Beach6_2019_calibration_data.csv')
```

```python
df
```

```python
raw_ecoli = 10 ** df['ECOLI_LOG10']
raw_ecoli
```

```python
plt.hist(raw_ecoli, bins=20, edgecolor='black')
plt.xlabel("E-Coli Level")
plt.ylabel("Frequency")
plt.title("Distribution of Raw E-Coli Levels")
plt.show()
```

```python
plt.hist(df['ECOLI_LOG10'], bins=20, edgecolor='black')
plt.xlabel("E-Coli Level")
plt.ylabel("Frequency")
plt.title("Distribution of Log10 E-Coli Levels")
plt.show()
```

```python
plt.scatter(df['TURB_NTRU'], df['ECOLI_LOG10'])
plt.xlabel("Turbidity")
plt.ylabel("ECOLI_LOG10")
plt.title("Turbidity vs. ECOLI_LOG10")
plt.show()
correlation_coefficient = np.corrcoef(df['TURB_NTRU'], df['ECOLI_LOG10'])
print(correlation_coefficient[0,1])
```

```python
Log10_Turbidity = np.log10(df['TURB_NTRU'])
plt.scatter(Log10_Turbidity, df['ECOLI_LOG10'])
plt.xlabel("Log10 Turbidity")
plt.ylabel("ECOLI_LOG10")
plt.title("Log10 Turbidity vs. ECOLI_LOG10")
plt.show()
correlation_coefficient = np.corrcoef(Log10_Turbidity, df['ECOLI_LOG10'])
print(correlation_coefficient[0,1])
```

```python
plt.scatter(df['RHUM_PCT'], df['ECOLI_LOG10'])
plt.xlabel("Relative Humidity(%)")
plt.ylabel("ECOLI_LOG10")
plt.title("Relative Humidity vs. ECOLI_LOG10")
plt.show()
correlation_coefficient = np.corrcoef(df['RHUM_PCT'], df['ECOLI_LOG10'])
print(correlation_coefficient[0,1])
```
